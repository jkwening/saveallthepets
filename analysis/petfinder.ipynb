```python
import os
import sys
import pandas as pd
import csv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Add project path to sys.path
ROOT_DIR = os.path.abspath('..')
if ROOT_DIR not in sys.path:
    sys.path.append(ROOT_DIR)
    
# Get DATA_DIR from project utils module
from utils import DATA_PATH

# Get petfinder files
petfinder_fnames = list()
for fname in os.listdir(DATA_PATH):
    if 'petfinder' in fname:
        petfinder_fnames.append(fname)
        
petfinder_fnames
```

```python
# Initialize dataframe
df = pd.read_csv(os.path.join(DATA_PATH, petfinder_fnames[0]))
df.tail()
```

```python
# Load all petfinder csv into single dataframe
for i, fname in enumerate(petfinder_fnames):
    time_stamp = fname.split('.csv')[0].split('_')[-1]
    if i == 0:
        df['time_stamp'] = time_stamp
        print(df.tail())
        continue
    
    temp_df = pd.read_csv(os.path.join(DATA_PATH, fname))
    temp_df['time_stamp'] = time_stamp
    df = df.append(temp_df, ignore_index=True)
    print(df.tail())
```
